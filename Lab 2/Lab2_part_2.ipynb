```
#Your Code 
import cv2
from skimage import io
from matplotlib import pyplot as plt
from skimage.io import imread, imsave, imshow

image = cv2.imread('noisy.jpg')

meadianFilter = cv2.medianBlur(image, 3)

gaussianFilter = cv2.GaussianBlur(image,(5,5),0)

plt.subplot(131), imshow(image), plt.title('Original')
plt.xticks([]), plt.yticks([])

plt.subplot(132), plt.imshow(meadianFilter), plt.title('Median Filter')
plt.xticks([]), plt.yticks([])

plt.subplot(133), plt.imshow(gaussianFilter), plt.title('Gaussian Filter')
plt.xticks([]), plt.yticks([])

#Median filter was most successful from removing the noise from the image
```
