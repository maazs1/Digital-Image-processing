```
#Your Code 
from skimage import io, img_as_ubyte
from skimage.filters import median, gaussian


damagePic= io.imread('damage_cameraman.png',as_gray=True)
damagePic= img_as_ubyte(damagePic, force_copy = False)
io.imshow(damagePic, cmap= "gray")
io.show()

damageMask = io.imread('damage_mask.png', as_gray=True)
damageMask= img_as_ubyte(damageMask, force_copy = False)
height, width = damagePic.shape

filterPic= gaussian(damagePic)
for x in range(0, height):
  for y in range (0, width):
    if damageMask[x][y] != 0:
      filterPic[x][y] = damagePic[x][y]

for x in range(0, 100):
  filterPic = gaussian(filterPic)
  for x in range(0, height):
    for y in range (0, width):
      if damageMask[x][y] != 0:
        filterPic[x][y] = damagePic[x][y]

io.imshow(filterPic, cmap='gray')
io.show()


```
