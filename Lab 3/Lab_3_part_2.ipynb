```

from matplotlib import pyplot as plt
from skimage import filters
import numpy as np
from skimage import io
from scipy.ndimage import gaussian_filter
from skimage import feature


image = io.imread('ex2.jpg')

gaussianF = gaussian_filter(image, sigma=1.0)


low25 = feature.canny(gaussianF, low_threshold=25, sigma=1)
plt.subplot(2, 2, 1)
plt.title("low threshold=25")
plt.imshow(low25, cmap='gray')
plt.xticks([]), plt.yticks([])

low50 = feature.canny(gaussianF, low_threshold=50, sigma=1)
plt.subplot(2, 2, 2)
plt.title("low threshold=50")
plt.imshow(low50, cmap='gray')
plt.xticks([]), plt.yticks([])

high150 = feature.canny(gaussianF, high_threshold=150, sigma=1)
plt.subplot(2, 2, 3)
plt.title("high threshold=150")
plt.imshow(high150, cmap='gray')
plt.xticks([]), plt.yticks([])

high200 = feature.canny(gaussianF, high_threshold=200, sigma=1)
plt.subplot(2, 2, 4)
plt.title("high threshold=200")
plt.imshow(high200, cmap='gray')
plt.xticks([]), plt.yticks([])
```

```
from matplotlib import pyplot as plt
from skimage import filters
import numpy as np
from skimage import io
from scipy.ndimage import gaussian_filter
from skimage import feature


image = io.imread('ex2.jpg')

gaussianF = gaussian_filter(image, sigma=1)


sigma1 = feature.canny(gaussianF, low_threshold=50, high_threshold=150, sigma=1.0)
plt.subplot(2, 2, 1)
plt.title("sigma=1")
plt.imshow(sigma1, cmap='gray')
plt.xticks([]), plt.yticks([])

sigma1_5 = feature.canny(gaussianF, low_threshold=50, high_threshold=150, sigma=1.5)
plt.subplot(2, 2, 2)
plt.title("sigma=1.5")
plt.imshow(sigma1_5, cmap='gray')
plt.xticks([]), plt.yticks([])

sigma2 = feature.canny(gaussianF, low_threshold=50, high_threshold=150, sigma=2.0)
plt.subplot(2, 2, 3)
plt.title("sigma=2.0")
plt.imshow(sigma2, cmap='gray')
plt.xticks([]), plt.yticks([])

sigma2_5 = feature.canny(gaussianF, low_threshold=50, high_threshold=150, sigma=2.5)
plt.subplot(2, 2, 4)
plt.title("sigma=2.5")
plt.imshow(sigma2_5, cmap='gray')
plt.xticks([]), plt.yticks([])



```
