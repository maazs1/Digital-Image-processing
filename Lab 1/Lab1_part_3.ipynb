```
#Your Code 
import cv2
import numpy as np
import matplotlib.pyplot as plt 
from skimage.io import imread, imsave, imshow
from skimage import io, img_as_ubyte, exposure
import math

def main(): 
  dayImage  = img_as_ubyte(io.imread('day.jpg', as_gray = True))
  nightImage = img_as_ubyte(io.imread('night.jpg', as_gray = True))

  dayHistogram = exposure.histogram(dayImage, nbins=256)
  plt.subplot(121),
  plt.title('Day Histogram')
  plt.xlim(0,256)
  plt.ylim(0,5500)
  plt.plot(dayHistogram[1], dayHistogram[0])

  nightHistogram = exposure.histogram(nightImage, nbins= 256)
  plt.subplot(122),
  plt.title('Night Histogram')
  plt.xlim(0,256)
  plt.ylim(0,5500)
  plt.plot(nightHistogram[1], nightHistogram[0])

  plt.show

  newDay = calculation(dayHistogram[0])
  newNight = calculation(nightHistogram[0])

  bahattacharya_C = 0

  for i in range(len(newDay)):
    bahattacharya_C +=math.sqrt((newDay[i]*newNight[i]))
  print("The Bahattacharya Coefficient is: ", bahattacharya_C )


  
def calculation(histogram):
  cumaltiveSum = 0
  for i in histogram:
    cumaltiveSum+=i
  newHistogram= np.zeros(256)
  
  for i in range(len(newHistogram)):
    newHistogram[i] = histogram[i]/cumaltiveSum

  return newHistogram 

if __name__== "__main__":
  main()

  



```
