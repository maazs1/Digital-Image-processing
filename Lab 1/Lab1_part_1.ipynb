```

import cv2
import numpy as np
import matplotlib.pyplot as plt 
from skimage.io import imread, imsave, imshow
from skimage import io, img_as_ubyte, exposure


im = img_as_ubyte(io.imread('test.jpg', as_gray=True))

#Manually getting the histogram 
y=np.zeros((256))
for i in range (im.shape[0]):
  for j in range (im.shape[1]):
    y[im[i,j]]+=1
x = np.arange(0,256)
plt.plot(x,y)
plt.show()

#Built in function to get histogram 
histr = exposure.histogram(im, nbins=256)
plt.xlim(0,256)
plt.plot(histr[1], histr[0])
plt.show




```
