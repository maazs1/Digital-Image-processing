```
import numpy as np
import matplotlib.pyplot as plt 
from skimage.io import imread, imsave, imshow
from skimage import io, img_as_ubyte, exposure

im = img_as_ubyte(io.imread('test.jpg', as_gray=True))

y=np.zeros(256)
for i in range (im.shape[0]):
  for j in range (im.shape[1]):
    y[im[i,j]]+=1

cumaltiveSum=np.cumsum(y)
for i in range (im.shape[0]):
  for j in range (im.shape[1]):
    im[i][j] = np.floor((((255)/((im.shape[0])*(im.shape[1])))*(cumaltiveSum[im[i][j]]))+0.5)

imshow(im)
plt.show()

newY=np.zeros((256))
for i in range (im.shape[0]):
  for j in range (im.shape[1]):
    newY[im[i,j]]+=1
newX = np.arange(0,256)
plt.plot(newX,newY)
plt.show()
```
